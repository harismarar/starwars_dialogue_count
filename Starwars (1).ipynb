## Text mining operations

```python

import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import pandas as pd


def process_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Tokenize text
    words = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalnum() and word not in stop_words]
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    return words


starwars_df['new_script'] = starwars_df['Dialogue'].apply(process_text)

print(starwars_df.head())
```

## Discovering the most relevent words

```python
# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



# Function to preprocess text
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Tokenize text
    words = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalnum() and word not in stop_words]
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(words)

# Combine the scripts and preprocess
starwars_df['clean_script'] = starwars_df['new_script'].apply(lambda x: ' '.join(x))

# TF-IDF Vectorization
vectorize_tfidf = TfidfVectorizer()
matrix_tfidf = vectorize_tfidf.fit_transform(starwars_df['clean_script'])

# Feature names
feature_names_tfidf = vectorize_tfidf.get_feature_names_out()

# Sum of the TF-IDF scores
sum_scores_tfidf = np.sum(matrix_tfidf.toarray(), axis=0)

# Words and their TF-IDF scores
tfidf_df = pd.DataFrame({'Words': feature_names_tfidf, 'tfidf_score': sum_scores_tfidf})

# Extract top n words by TF-IDF score
top_n_words = tfidf_df.sort_values(by='tfidf_score', ascending=False).head(30)
print("Top words by TF-IDF score:")
print(top_n_words)

# Plot the most relevant words
plt.figure(figsize=(12, 6))
plt.bar(top_n_words['Words'], top_n_words['tfidf_score'])
plt.title('Most Relevant Words in The Original Trilogy Scripts')
plt.xlabel('Words')
plt.ylabel('TF-IDF Score')
plt.xticks(rotation=90)
plt.show()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk



# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Add columns to store sentiment scores
starwars_df["compound_score"] = starwars_df["Dialogue"].apply(lambda x: sia.polarity_scores(x)["compound"])
starwars_df["positive_score"] = starwars_df["Dialogue"].apply(lambda x: sia.polarity_scores(x)["pos"])
starwars_df["negative_score"] = starwars_df["Dialogue"].apply(lambda x: sia.polarity_scores(x)["neg"])
starwars_df["neutral_score"] = starwars_df["Dialogue"].apply(lambda x: sia.polarity_scores(x)["neu"])

# Classify sentiments based on compound scores
starwars_df["sentiment"] = starwars_df["compound_score"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))

# Display the dataframe with sentiment analysis results
print(starwars_df.head())
```

```python
# Calculate overall sentiment for each episode
total_average_sentiment = starwars_df.groupby("Episode").agg({
    "compound_score": "mean",
    "positive_score": "mean",
    "negative_score": "mean",
    "neutral_score": "mean"
}).reset_index()

# Determine the predominant sentiment for each episode
total_average_sentiment["predominant_sentiment"] = total_average_sentiment["compound_score"].apply(
    lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral")
)

# Display overall sentiment for each episode
print(total_average_sentiment)
```
